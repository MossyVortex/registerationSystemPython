```python
def tenMostAbcentStudents(information):
    
    outputFile=open("arranged_students_record.txt","w")
    outputFile.write("ID                 NAME              Abcences             Exam1 Grade             Exam2 Grade             Total Marks\n")
    outputFile.write("-"*120+ "\n")
    holderForName=[]
    x=0
    #for calculating how many student will add.
    for key in information:
        holderForName.append(int(information[key][1]))
    holderForName.sort(reverse=True)
    #for sorting the abcences  from the larger to lower.
    nameList=[]
    for item in holderForName:
        if x<10:
            x=x+1
            if item not in nameList:
                nameList.append(item)
    nameSorted={}
    for item in nameList:
        for key in information:
            if str(item)==information[key][1]:
                nameSorted[key]=information[key]
    for key in nameSorted:
        ID=key
        name_sort=nameSorted[key]
        name=name_sort[0]
        abcence=name_sort[1]
        exam1Grade=name_sort[2]
        exam2Grade=name_sort[3]
        totalMark=name_sort[4]
        #saving student information into varuble.
        outputFile.write("{0:13} {1:25} {2:22} {3:22} {4:13}          {5:13} \n".format(ID,name,abcence,exam1Grade,exam2Grade,totalMark))
        #generating the information to the file.
    print("Showing the ten most absent students has done")
    
def sortedTotalMarks(information):# the function that sorting the information according to total marks.
    
    outputFile=open("arranged_students_record.txt","w")
    outputFile.write("ID                 NAME              Abcences             Exam1 Grade             Exam2 Grade             Total Marks\n")
    outputFile.write("-"*120+ "\n")
    holderForName=[]
    for key in information:
        holderForName.append(int(information[key][4]))
    holderForName.sort()
    nameList=[]
    for item in holderForName:
        if item not in nameList:
            nameList.append(item)
    nameSorted={}
    for item in nameList:
        for key in information:
            if str(item)==information[key][4]:
                nameSorted[key]=information[key]
    # this all for sorting to the forth valu in dictionry.
    for key in nameSorted:
        ID=key
        name_sort=nameSorted[key]
        name=name_sort[0]
        abcence=name_sort[1]
        exam1Grade=name_sort[2]
        exam2Grade=name_sort[3]
        totalMark=name_sort[4]
        #saving student information into varuble.
        outputFile.write("{0:13} {1:25} {2:22} {3:22} {4:13}          {5:13} \n".format(ID,name,abcence,exam1Grade,exam2Grade,totalMark))
        #generating the information to the file.
    print("Sorting by total marks has done")  
    
def sortedName(information): # the function that sorting the information according to names.
    
    outputFile=open("arranged_students_record.txt","w")
    outputFile.write("ID                 NAME              Abcences             Exam1 Grade             Exam2 Grade             Total Marks\n")
    outputFile.write("-"*120+ "\n")
    #the frame
    information=dict(sorted(information.items(),key=lambda x:x[1]))
    #to sort names which is represent values.
    for key in information :
        ID=key
        inform=information[key]
        name=inform[0]
        abcence=inform[1]
        exam1Grade=inform[2]
        exam2Grade=inform[3]
        totalMark=int(exam1Grade)+int(exam2Grade)
        #saving student information into varuble.
        outputFile.write("{0:13} {1:25} {2:22} {3:22} {4:13} {5:13} \n".format(ID,name,abcence,exam1Grade,exam2Grade,totalMark))
        #generating the information to the file.
    print("Sorting by name has done")
    
def sortedID(information): # the function that sorting the information according to ID.
    
    outputFile=open("arranged_students_record.txt","w")
    outputFile.write("ID                 NAME              Abcences             Exam1 Grade             Exam2 Grade             Total Marks\n")
    outputFile.write("-"*120+ "\n")
    #the frame
    for key in sorted(information):# to visit each valuse in infrmation and sorted it.
        ID=key
        inform=information[key]
        name=inform[0]
        abcence=inform[1]
        exam1Grade=inform[2]
        exam2Grade=inform[3]
        totalMark=int(exam1Grade)+int(exam2Grade)
        #saving student information into variable.
        outputFile.write("{0:13} {1:25} {2:22} {3:22} {4:13} {5:13} \n".format(ID,name,abcence,exam1Grade,exam2Grade,totalMark))
        #generating the information to the file.
    print("Sorting by ID has done")
    
def sortedAbcences(information): # the function that sorting the information according to Abcences.
    
    outputFile=open("arranged_students_record.txt","w")
    outputFile.write("ID                 NAME              Abcences             Exam1 Grade             Exam2 Grade             Total Marks\n")
    outputFile.write("-"*120+ "\n")
    holderForName=[]
    for key in information:
        holderForName.append(int(information[key][1]))
    holderForName.sort()
    nameList=[]
    for item in holderForName:
        if item not in nameList:
            nameList.append(item)
    nameSorted={}
    for item in nameList:
        for key in information:
            if str(item)==information[key][1]:
                nameSorted[key]=information[key]
    # this all for sorting to the firs valu in dictinery.
    for key in nameSorted:
        ID=key
        name_sort=nameSorted[key]
        name=name_sort[0]
        abcence=name_sort[1]
        exam1Grade=name_sort[2]
        exam2Grade=name_sort[3]
        totalMark=name_sort[4]
        #saving student information into varuble.
        outputFile.write("{0:13} {1:25} {2:22} {3:22} {4:13}          {5:13} \n".format(ID,name,abcence,exam1Grade,exam2Grade,totalMark))
        #generating the information to the file.
    print("Sorting by absences has done")
    
def topTenStudents(information):# the function that show top ten students.
    
    outputFile=open("arranged_students_record.txt","w")
    outputFile.write("ID                 NAME              Abcences             Exam1 Grade             Exam2 Grade             Total Marks\n")
    outputFile.write("-"*120+ "\n")
    holderForName=[]
    x=0
    #for calculating how many student will add.
    for key in information:
        holderForName.append(int(information[key][4]))
    holderForName.sort(reverse=True)
    #for sorting the total marks from the larger to lower.
    nameList=[]
    for item in holderForName:
        if x<10:
            x=x+1
            if item not in nameList:
                nameList.append(item)
    nameSorted={}
    #save the information in dict.
    for item in nameList:
        for key in information:
            if str(item)==information[key][4]:
                nameSorted[key]=information[key]
    for key in nameSorted:
        ID=key
        name_sort=nameSorted[key]
        name=name_sort[0]
        abcence=name_sort[1]
        exam1Grade=name_sort[2]
        exam2Grade=name_sort[3]
        totalMark=name_sort[4]
        #saving student information into varuble.
        outputFile.write("{0:13} {1:25} {2:22} {3:22} {4:13}          {5:13} \n".format(ID,name,abcence,exam1Grade,exam2Grade,totalMark))
        #generating the information to the file.
    print("Showing top 10 students has done")
    
def main2(sorting):

    # initialization and opening files
    textFile=open("students_records.txt","r")
    outputFile=open("arranged_students_record.txt","w")
    information = {}

    # information holder and names.
    for line in textFile:
            wordList = line.split()
            counter = 0
            for words in wordList:
                if counter == 0:
                    wordList.pop(0)
                    information[words] = wordList
                counter += 1
    textFile.close()           
    outputFile.close()

    # to split words indivsiualy.
    for key in information:

            ID=key
            inform=information[key]
            name=inform[0]
            abcence=inform[1]
            exam1Grade=inform[2]
            exam2Grade=inform[3]
            totalMark=int(exam1Grade)+int(exam2Grade) #the total mark calculation.
            inform.insert(4,str(totalMark))

    # executing choice
    if sorting=="a":
            sortedID(information)
            textFile.close()
            outputFile.close()
    elif sorting=="b":
        sortedName(information)
        textFile.close()
        outputFile.close()
    elif sorting=="c":
        sortedTotalMarks(information)
        textFile.close()
        outputFile.close()
    elif sorting=="d":
        sortedAbcences(information)
        textFile.close()
        outputFile.close()
    elif sorting=="e":
        topTenStudents(information)
        textFile.close()
        outputFile.close()
    elif sorting=="f":
        tenMostAbcentStudents(information)
        textFile.close()
        outputFile.close()
        
def wipe():
    
    records_rewriteW = open("students_records.txt", "w")
    records_rewriteW.write("\n")
    records_rewriteW.close()
    
def rewrite(ID, name, absences, grade1, grade2):
    
    records_rewriteA = open("students_records.txt", "a")
    records_rewriteA.write("%s %s %s %s %s\n"%(ID, name, absences, grade1, grade2))
    records_rewriteA.close()
    
def cleanFile(): 
    
    records = open("students_records.txt", "r")

    IDS_list = []
    names_list = []
    absences_list = []
    grade1_list = []
    grade2_list = []

    for line in records:
        if str(line) != "\n":
            temp = line.split()
            IDS_list.append(temp[0])
            names_list.append(temp[1])
            absences_list.append(temp[2])
            grade1_list.append(temp[3])
            grade2_list.append(temp[4])
            
    records.close()
    
    records = open("students_records.txt", "w")
    
    for i in range(len(IDS_list)):
        rewrite(IDS_list[i], names_list[i], absences_list[i], grade1_list[i], grade2_list[i])

def add(ID, name, absences, grade1, grade2):
    
    recordsW = open("students_records.txt", "a")
    recordsW.write("\n%s %s %s %s %s"%(ID, name, absences, grade1, grade2))
    IDS_list.append(ID)
    recordsW.close()
    cleanFile()
    print("Student has been added successfully.")
    
def remove(ID):
    
    temp_index = IDS_list.index(ID)
    IDS_list.pop(temp_index)
    names_list.pop(temp_index)
    absences_list.pop(temp_index)
    grade1_list.pop(temp_index)
    grade2_list.pop(temp_index)
      
    wipe()
    
    for r in range(len(IDS_list)):
        rewrite(IDS_list[r], names_list[r], absences_list[r], grade1_list[r], grade2_list[r])
    cleanFile()
    print("the student with ID: %s has been removed successfully."%ID)
    
def modify(LIST, index, new_value):
    
    LIST.remove(LIST[index])
    LIST.insert(index, new_value)

IDS_list = []
names_list = []
absences_list = []
grade1_list = []
grade2_list = []
        
def main():    
        
    run_flag = True
    while run_flag:
        
        IDS_list.clear()
        names_list.clear() 
        absences_list.clear() 
        grade1_list.clear() 
        grade2_list.clear()
        
        records = open("students_records.txt", "r")
        
        cleanFile() #remove empty lines 

        for line in records:
            temp = line.split()
            IDS_list.append(temp[0])
            names_list.append(temp[1])
            absences_list.append(temp[2])
            grade1_list.append(temp[3])
            grade2_list.append(temp[4])

        print("Menu\n1. Add new student\n2. Remove a student\n3. Modify student info\n4. Show all students\n5. Show top 10 students\n6. Show most absent\n7. Exit")

        temp_user_input = input("Enter the number of your choice: ")

        ### chioce validation ###

        while not temp_user_input.isdigit() or not 0 < int(temp_user_input) < 8:
            temp_user_input = input("Wrong input! enter a number from 1-7: ")

        ### end of choice validation ###


        ### excution of chioce ###

        if temp_user_input == '1':

            ID = input("Enter the ID you want to add: ") 

            ### ID validation ###

            flag = True
            while flag:

                if IDS_list.count(ID) > 0:
                    ID = input("This ID already exists, choose another one: ")

                else:
                    if len(ID) != 9:
                        ID = input("This ID is either short or long, enter an ID that contains 9 digits only: ")

                    elif not ID.isdigit():
                        ID = input("Error: ID should only contain 9 digits, enter another ID: ")

                    else:
                        flag = False

            ### end of ID validation ###
            
            ### name input and validation ###

            first = input("Enter first name: ")

            while not first.isalpha() or ' ' in first:
                first = input("invalid name, enter a name that contains letters only: ")

            last = input("Enter last name: ")    

            while not last.isalpha() or ' ' in last:
                last = input("invalid name, enter a name that contains letters only: ")

            name = first + "-" + last
            
            ### end of name input ###
            
            ### absences input and validation ###
            
            absences = input("Enter the student's absences: ")
            
            while not absences.isdigit():
                absences = input("Error: absences should be a number, enter absences: ")

            ### end of absences input and validation ###
            
            ### grades input and validation ###
            
            exam1_grade = input("Enter the student's exam 1 grade: ")

            while not exam1_grade.isdigit() or not (0 <= int(exam1_grade) <= 100):
                exam1_grade = input("Error: exam grades should only be a number between 0 and 100, enter grade1: ")
                
            exam2_grade = input("Enter the student's exam 2 grade: ")
            
            while not exam2_grade.isdigit() or not (0 <= int(exam2_grade) <= 100):
                exam2_grade = input("Error: exam grades should only be a number between 0 and 100, enter grade2: ")
            
            ### end of grades input and validation ###
            
            add(ID, name, absences, exam1_grade, exam2_grade)

        elif temp_user_input == '2':

            ID = input("Enter the ID you want to remove: ")

            ### ID validation ###

            flag = True
            while flag:

                if IDS_list.count(ID) == 0:
                    ID = input("This ID does not exist, choose another one: ")

                else:
                    if len(ID) != 9:
                        ID = input("This ID is either short or long, enter an ID that contains 9 digits only: ")

                    elif not ID.isdigit():
                        ID = input("Error: ID should only contain 9 digits, enter another ID: ")

                    else:
                        flag = False

            ### end of ID validation ###            

            remove(ID)

        elif temp_user_input == '3':

            ID = input("Enter the ID you want to modify: ")

            ### ID validation ###

            flag = True
            while flag:

                if IDS_list.count(ID) == 0:
                    ID = input("This ID does not exist, choose another one: ") # user can not modify an ID that does not exist

                else:
                    if len(ID) != 9:
                        ID = input("This ID is either short or long, enter an ID that contains 9 digits only: ")

                    elif not ID.isdigit():
                        ID = input("Error: ID should only contain 9 digits, enter another ID: ")

                    else:
                        flag = False

            ### end of ID validation ###

            temp_index = IDS_list.index(ID)

            print("1." + IDS_list[temp_index] + "\n2." + names_list[temp_index] + "\n3." + absences_list[temp_index] + "\n4." + grade1_list[temp_index] + "\n5." + grade2_list[temp_index])

            choice = int(input("Enter the number of the element you want to modify: "))

            ### choice validation ###
            
            choice_str = str(choice)
            while not choice_str.isdigit() and not 1 <= choice < 6:
                choice = input("Wrong input, enter a number between 1 and 5: ")

            ### end of choice validation ###

            if choice == 1: # add new student
                
                new = input("Enter the new value: ")
                
                ### ID validation ###
                
                flag = True
                while flag:

                    if IDS_list.count(new) > 0:
                        new = input("This ID already exists, choose another one: ") # when modifying ID user should not dublicate old one

                    else:
                        if len(new) != 9:
                            new = input("This ID is either short or long, enter an ID that contains 9 digits only: ")

                        elif not new.isdigit():
                            new = input("Error: ID should only contain 9 digits, enter another ID: ")

                        else:
                            flag = False
                    
                ### end of ID validation ###
                
                modify(IDS_list, temp_index, new)

            elif choice == 2: # remove current student
                
                first = input("Enter first name: ")

                while not first.isalpha() or ' ' in first:
                    first = input("invalid name, enter a name that contains letters only: ")

                last = input("Enter last name: ")    

                while not last.isalpha() or ' ' in last:
                    last = input("invalid name, enter a name that contains letters only: ")

                new = first + "-" + last
                
                modify(names_list, temp_index, new)

            elif choice == 3: # modify current student
                
                ### absences validation ###

                new = input("Enter the student's absences: ")
            
                while not new.isdigit():
                    new = input("Error: absences should be a number, enter absences: ")
                
                ### end of absences valiation ###
                
                modify(absences_list, temp_index, new)

            elif choice == 4:
                
                new = input("Enter the student's exam 1 grade: ")
            
                while not new.isdigit() or not (0 <= int(new) <= 100):
                    new = input("Error: exam grades should only be a number between 0 and 100, enter grade1: ")
                
                modify(grade1_list, temp_index, new)

            elif choice == 5:
                
                new = input("Enter the student's exam 2 grade: ")
            
                while not new.isdigit() or not (0 <= int(new) <= 100):
                    new = input("Error: exam grades should only be a number between 0 and 100, enter grade2: ")

                modify(grade2_list, temp_index, new)

            wipe() #wipe the output file, but the values are saved.
            
            for t in range(len(IDS_list)):
                rewrite(IDS_list[t], names_list[t], absences_list[t], grade1_list[t], grade2_list[t]) #rewrite values in the output file

            cleanFile() #get rid of empty lines

            print("The value was modified successfully.")

        elif temp_user_input == '4': # show all students 
            print("a. sort by ID.\nb. sort by name.\nc. sort by total marks.\nd. sort by abcences.")
            sorting=input("Enter a, b, c or d:")
            
            if sorting=="a":
                main2('a')
            elif sorting=="b":
                main2('b')
            elif sorting=="c":
                main2('c')
            elif sorting=="d":
                main2('d')
            
        elif temp_user_input == '5': # show top 10 students
            main2('e')
            
        elif temp_user_input == '6': # show students with largest absences
            main2('f')
            
        elif temp_user_input == '7': # terminate program
            run_flag = False
            print("closing.")

        ### end of excuting choice ### 

        records.close()

main()
```
